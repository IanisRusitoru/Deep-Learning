```python
import zipfile
import os
import numpy as np
from PIL import Image
%matplotlib inline
import matplotlib.pyplot as plt
import cv2
```

__`Step 1`__ Unzip data.zip

```python
# Define the path to the zip file
zip_file_path = 'C:\\Users\\Chip7 MemMartins\\Desktop\\Project\\data.zip'

# Define the path to the directory where you want to extract the files
extract_dir_path = 'C:\\Users\\Chip7 MemMartins\\Desktop\\Project'

# Extract the zip file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extract_dir_path)

# Print the list of extracted files
print(f"The files have been extracted to {extract_dir_path}:")
for root, dirs, files in os.walk(extract_dir_path):
    for file in files:
        print(os.path.join(root, file))
```

__`Step 2`__ Properties of the data

```python
# Define the path to the dataset folder
data_dir = "C:\\Users\\Chip7 MemMartins\\Desktop\\Project\\data"

# Get the list of classes (i.e., subfolders) in the data directory
classes = os.listdir(data_dir)

num_images = []
total_images = 0
image_sizes = []

for c in classes:
    class_dir = os.path.join(data_dir, c)
    # Get the list of image file names in the class folder
    file_names = os.listdir(class_dir)
    num_images.append(len(file_names))
    total_images += len(file_names)
    # Loop through each image file in the class folder
    for f in file_names:
        # Open the image and get its size
        img = Image.open(os.path.join(class_dir, f))
        size = img.size
        image_sizes.append(size)

image_sizes = np.array(image_sizes)

print(data_dir)
print("Number of classes:", len(classes))
for i, c in enumerate(classes):
    print("Number of images in", c, ":", num_images[i])
print("Total number of images:", total_images)
print("Size of images (width x height):", np.unique(image_sizes, axis=0))
```

```python
def get_aspect_ratios(folder_path):
    aspect_ratios = []
    for image_name in os.listdir(folder_path):
        image_path = os.path.join(folder_path, image_name)
        img = cv2.imread(image_path)
        if img is not None:
            height, width, _ = img.shape
            aspect_ratio = width / height
            aspect_ratios.append(aspect_ratio)
    return aspect_ratios
```

```python
#cardboard choose as an example
aspect_ratios = get_aspect_ratios("C:\\Users\\Chip7 MemMartins\\Desktop\\Project\\data\\cardboard")

plt.hist(aspect_ratios, bins=20)
plt.title('Aspect Ratio Histogram')
plt.xlabel('Aspect Ratio')
plt.ylabel('Frequency')
plt.show()
```

__`Step 3`__ Data Visualization

```python
# Define the path to the dataset folder
data_dir = 'C:\\Users\\Chip7 MemMartins\\Desktop\\Project\\data'

image_paths = []
labels = []

# Loop over each sub-folder
for subdir in os.listdir(data_dir):
    subdir_path = os.path.join(data_dir, subdir)
    if os.path.isdir(subdir_path):
        # Get a list of all image filenames in the sub-folder
        image_filenames = [os.path.join(subdir_path, filename) for filename in os.listdir(subdir_path) if filename.endswith(".jpg")]

        # Add the image filenames to the list
        image_paths += image_filenames
        # Add the label for each image to the list
        labels += [subdir] * len(image_filenames)

# Display two rows of images for each class
classes = set(labels)
num_classes = len(classes)

fig, axes = plt.subplots(num_classes*2, 10, figsize=(20, 6*num_classes))

for i, cls in enumerate(classes):
    class_images = [img for img, lbl in zip(image_paths, labels) if lbl == cls]
    selected_indices = np.random.choice(len(class_images), size=20, replace=False)
    row1 = i*2
    row2 = i*2+1
    axes[row1, 0].set_title(cls)
    axes[row1, 0].axis('off')
    axes[row2, 0].axis('off')
    for j, idx in enumerate(selected_indices):
        with Image.open(class_images[idx]) as img:
            col = j%10
            if j < 10:
                axes[row1, col].imshow(img)
                axes[row1, col].axis('off')
            else:
                axes[row2, col].imshow(img)
                axes[row2, col].axis('off')

plt.show()
```

__`Step 4`__ Decision

#### Eliminiation of class 'Trash'
After visual inspection we arrived to the conclusion that 'trash' does not have unique characteristics compared to other classes, that is, that would distinguish them from the others. In this way, this class is constituted by parts of the remaining classes and therefore we decided to eliminate it (i.e: plastic, paper). Moreover, it's specified in git-hub that some of the images in 'trash' are mislabeled.
